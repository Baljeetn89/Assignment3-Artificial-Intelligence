Consider the data from the file weight-height.csv.

1- Inspect the dependence between height and weight using a scatter plot. You may use either of the variables as independent variable.

2- Choose appropriate model for the dependence

3- Perform regression on the data using your model of choice

It is recommended that you use the module sklearn for all your computations.

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the values of n
n_values = [500, 1000, 2000, 5000, 10000, 15000, 20000, 50000, 100000]

for n in n_values:
    # Step 2: Simulate throwing two dice n times and compute the sum
    s = np.random.randint(1, 7, size=n) + np.random.randint(1, 7, size=n)
    
    # Step 3: Compute the histogram of sums
    h, h2 = np.histogram(s, bins=range(2, 14))
    
    # Step 4: Plot the normalized histogram
    plt.figure(figsize=(8, 5))
    plt.bar(h2[:-1], h / n, color='skyblue')
    plt.title(f'Histogram of Dice Sums for n = {n}')
    plt.xlabel('Sum of Dice')
    plt.ylabel('Frequency')
    plt.ylim(0, 0.2)
    plt.show()
```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt

# Step 1: Load the data
data = pd.read_csv('weight-height.csv')  # Ensure the file is in the correct location

# Step 2: Create a scatter plot
plt.scatter(data['Height'], data['Weight'], alpha=0.5)
plt.title("Scatter Plot of Height vs. Weight")
plt.xlabel("Height (inches)")
plt.ylabel("Weight (pounds)")
plt.grid()
plt.show()

# Step 3: Choose model and perform regression
X = data[['Height']]  # Independent variable
y = data['Weight']     # Dependent variable

model = LinearRegression()
model.fit(X, y)

# Step 4: Plot regression results
plt.scatter(data['Height'], data['Weight'], alpha=0.5)
plt.plot(data['Height'], model.predict(X), color='red', linewidth=2)  # Regression line
plt.title("Height vs. Weight with Regression Line")
plt.xlabel("Height (inches)")
plt.ylabel("Weight (pounds)")
plt.grid()
plt.show()

# Step 5: Compute RMSE and R
y_pred = model.predict(X)
rmse = np.sqrt(mean_squared_error(y, y_pred))
r2 = r2_score(y, y_pred)

print(f"RMSE: {rmse:.2f}")
print(f"R: {r2:.2f}")

# Step 6: Quality assessment
print("The RMSE value indicates the average error in weight predictions, while R indicates the proportion of variance explained by the model.")
```
